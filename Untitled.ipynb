```python


# from PIL import Image as PILImage
import matplotlib.pyplot as plt
from  io import BytesIO
import requests
# import re
from bs4 import BeautifulSoup
from IPython.display import Image, display

headers = {
   "User-Agent":  "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 Edg/112.0.1722.58",
    "Referer": "https://movie.douban.com/top250"
 }

page_step = 25

def find_depth(res):
    soup = BeautifulSoup(res.text,'html.parser')
    depth = soup.find('span',class_='next').previous_sibling.previous_sibling.text
    return int(depth)

def open_url(url):
    res = requests.get(url, headers=headers)
    return res

def find_movies(url):
    response = requests.get(url, headers=headers, )
    if response.status_code == 200:
        soup = BeautifulSoup(response.text,'html.parser')
        movie_list = []
        for movie in soup.find_all('div',{'class': 'info'}):
            try:
                img_url=movie.parent.find('img')['src']
                response=requests.get(img_url,timeout=10)
                # img=PILImage.open(BytesIO(response.content))

                title = movie.find('span',{'class': 'title'}).text
                other_title = movie.find('span',{'class': 'other'}).text
                rating = movie.find('span',{'class': 'rating_num'}).text
                quote = movie.find('span',{'class': 'inq'}).text if movie.find('span',{'class':'inq'}) else ''


                movie_list.append({
                    'title': title,
                 'other_title': other_title,
                    'rating': rating,
                    'quote': quote,
                    'img_url': img_url,
                    # 'img': img
                
              })
            except requests.exceptions.Timeout:
                    print(f"picture was ovetime:{img_url}")
        return movie_list
    else:
        print('Failed')
        return []

print("it my favourite movies")



def main():
    host = "https://movie.douban.com/top250"
    res = open_url(host + "?start=0&filter=")
    depth = find_depth(res)
    print(depth)

    for page in range(0, depth * 25, 25):
        url = f"{host}?start={page}&filter="
        movie_list = find_movies(url)
        for movie in movie_list:
            print(movie)
            response= requests.get(movie['img_url'])
            img= plt.imread(BytesIO(response.content), format= 'jpg')
            plt.figure(figsize= (1, 1.5))
            plt.imshow(img)
            plt.axis('off')
            plt.show()
#             display(Image(url=movie['img_url'], width= 100))
            # display(Image(url=movie['img_url']))

main()




print ("no")
```
